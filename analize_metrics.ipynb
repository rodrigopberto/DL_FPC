```python
import pandas as pd
```

```python
df = pd.read_csv("sim_scores.csv")
df["snr"]=df["snr"].astype(str)
```

```python
df["freq_mae"]=df["abs_freq_error"]
df["freq_std"]=df["abs_freq_error"]
df["phase_mae"]=df["abs_phase_error"]
df["phase_std"]=df["abs_phase_error"]

df[df["wsup"]=="sup"].groupby(["snr","model","wsup"]).agg({"freq_mae":"mean","freq_std":"std","phase_mae":"mean","phase_std":"std","abs_freq_error":"count"})
```
